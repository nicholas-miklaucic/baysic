```python
from mp_api.client import MPRester
from rich.progress import track

df = []

filtered = [x for x in comps if x.num_atoms <= 20]
mp_ids = []
with MPRester(mute_progress_bars=True) as mpr:
    for comp in track(filtered):
        docs = mpr.summary.search(formula=comp.to_pretty_string())
        mp_ids.append(docs[0].material_id)
```

```python
pd.Series(list(map(str, mp_ids))).to_csv('test_mpids.csv')
```
